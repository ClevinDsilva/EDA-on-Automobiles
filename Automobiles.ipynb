```python
plt.figure(figsize=(10, 5))
sns.histplot(x="automobile_interested", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="primary_purpose", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="trusted_brand", hue="preferred_usage", palette="viridis")
plt.title("Preferred Automobile Usage vs. Trusted Brand (Bivariate)")
plt.xlabel("Trusted Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder


df_copy = df.copy()


categorical_columns = ["trusted_brand", "preferred_usage", "automobile_interested"]


label_encoder = LabelEncoder()
for col in categorical_columns:
    df_copy[col] = label_encoder.fit_transform(df_copy[col])


plt.figure(figsize=(8, 5))
sns.heatmap(df_copy[categorical_columns].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap: Trusted Brand, Preferred Usage & Automobile Interest (Multivariate)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="automobile_interested", y="financing_options", data=df)
plt.title("Financing Options for Different Automobile Types")
plt.xlabel("Type of Automobile Interested")
plt.ylabel("Preferred Financing Option")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["automobile_interested"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Automobile Interest (Univariate)")
plt.xlabel("Preferred Automobile Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
if df["trusted_brand"].dtype == "int64":
    brand_mapping = dict(enumerate(label_encoder.classes_))
    df["trusted_brand"] = df["trusted_brand"].map(brand_mapping)

pd.crosstab(df["trusted_brand"], df["preferred_usage"]).plot(kind="bar", stacked=True, figsize=(8, 5))
plt.title("Trusted Automobile Brands by Preferred Usage")
plt.xlabel("Trusted Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="new_or_preowned", hue="preferred_usage", palette="viridis")
plt.title("Preference for New vs. Preowned Automobiles (Bivariate)")
plt.xlabel("Purchase Preference (New or Preowned)")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x="tech_features_importance", y="automobile_interested", data=df)
plt.title("Tech Features Importance vs. Automobile Interest")
plt.xlabel("Tech Features Importance")
plt.ylabel("Type of Automobile Interested")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="preferred_usage", y="safety_features", data=df, palette="Set3")
plt.title("Safety Features vs. Preferred Usage", fontsize=16)
plt.xlabel("Preferred Usage", fontsize=14)
plt.ylabel("Safety Features Importance", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bubble_sizes = df["automobile_interested"].astype("category").cat.codes * 100
scatter = plt.scatter(df["storage_importance"], df["environmental_factors"],
                      s=bubble_sizes, alpha=0.5, c=bubble_sizes, cmap="coolwarm", label="Automobile Interest")
plt.xlabel("Storage Importance")
plt.ylabel("Environmental Factors")
plt.title("Bubble Chart: Storage Importance vs. Environmental Factors vs. Automobile Interest")
plt.colorbar(label="Automobile Interest Level")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df["new_or_preowned"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel", 5))
plt.ylabel("")
plt.title("New vs. Preowned Vehicle Preference")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

crosstab_data = pd.crosstab(df["financing_options"], df["preferred_engine"], normalize="index")

crosstab_data.plot(kind="bar", stacked=True, colormap="viridis", figsize=(10, 6))

plt.xlabel("Financing Options")
plt.ylabel("Proportion")
plt.title("Financing Options vs. Preferred Engine Type")
plt.legend(title="Engine Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="new_or_preowned", y="storage_importance", data=df, palette="viridis")
plt.title("Violin Plot: Storage Importance Based on New vs. Preowned Vehicles")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x="automobile_interested", y="primary_purpose", data=df, palette="coolwarm", dodge=True)
plt.title("Preferred Automobile Type by Primary Purpose", fontsize=16)
plt.xlabel("Automobile Type", fontsize=14)
plt.ylabel("Primary Purpose", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
